```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.models import model_from_json
%matplotlib inline
```

```python
path = './tests/test_1/'
test_df = pd.read_csv(path + 'training_history.csv', index_col=0)
history = test_df.loc['val_loss'].apply(lambda x: np.asarray(x[1:-1].split(','), dtype='float'))
with open(path + 'model_architecture_relu.txt', 'r') as f:
    model = model_from_json(f.read())

model.summary()
plt.plot(history['relu'])
plt.plot(history['sigmoid'])
plt.plot(history['tanh'])
plt.plot(history['sin'])
plt.legend(['relu', 'sigmoid', 'tanh', 'sin'])
plt.xlabel('Epochs')
plt.ylabel('Validation Loss')
plt.title("Comparing Different Activation Functions")
plt.show()
```
