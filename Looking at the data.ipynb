```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
test_df = pd.read_csv('./tests/test_1/training_history.csv', index_col=0)
history = test_df.loc['val_loss'].apply(lambda x: np.asarray(x[1:-1].split(','), dtype='float'))

plt.plot(history['relu'])
plt.plot(history['sigmoid'])
plt.plot(history['tanh'])
plt.plot(history['sin'])
plt.legend(['relu', 'sigmoid', 'tanh', 'sin'])
plt.show()
```

It's pretty clear Sin was the worst for this architecture.
